```python
df_sup = spark.table("PharmaLakehouse.bronze_suppliers")
display(df_sup)
```

```python
from pyspark.sql.functions import *
from pyspark.sql import *

df_sup = df_sup.dropDuplicates(["supplier_id"])
display(df_sup)
```

```python
df_sup.write.format("delta").mode("overwrite").saveAsTable("silver_suppliers")
```

# ORDERS TABLE

```python
df_orders = spark.table("PharmaLakehouse.bronze_raw_orders")
display(df_orders)
```

```python
df_orders = df_orders.dropDuplicates(["order_id"])
display(df_orders)
```

```python
df_orders = df_orders.dropna(subset=["order_id","material"])
display(df_orders)
```

```python
df_orders.write.format("delta").mode("overwrite").saveAsTable("silver_orders")
```
